# 1. Find out how many males and females participated in the test

```python
gender = data['gender'].value_counts().plot(kind='bar',title='Bar Graph of Gender',figsize=(7, 6),color='c')
plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.show()
```

# 2.The students' parental level of education

```python
df1 = data.groupby(['gender','parental level of education'])[['reading score', 'math score', 'writing score']].mean()
df1
```

# 3.Scores the most on average for math, reading and writing based on gender

```python
data['average score'] = (data['math score'] + data['writing score'] + data['reading score']) / 3
data['average score']
```

```python
df2 = data.groupby(['parental level of education','test preparation course'])[['reading score', 'math score', 'writing score']].mean()
df2
```

```python
df3 = data.groupby(['gender','test preparation course'])[['reading score', 'math score', 'writing score']].mean()
df3
```

```python
data[data['average score'] == 100]
average_score_data = data[data['average score'] >= 90][data.columns]
average_score_data
```

# 4. Scoring variation for math reading and writing based on gender and test preparation course
